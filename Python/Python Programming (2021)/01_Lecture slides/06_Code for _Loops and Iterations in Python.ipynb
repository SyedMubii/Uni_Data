```python
while True: 
    user_input = input('Say something: ')
    if user_input[0] == '#': continue
    if user_input == 'done': break 
    print('You input was: ' + user_input)
print('Good bye!')
```

```python
my_list = [9, 41, 12, 3, 74, 15]

largest_so_far = -1

for i in my_list: 
    if i > largest_so_far: 
        largest_so_far = i

print(f"The largest number is {largest_so_far}")
```

```python
counter = 0
total = 0

for i in my_list: 
    counter = counter + 1
    total = total + i

print("Average = {}".format(total / counter))
```
